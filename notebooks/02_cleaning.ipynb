```python
import pandas as pd
import sys

sys.path.insert(0,'../scripts')
from db_connect import get_connection

engine_df = pd.read_excel('../data/games_data.xlsx',sheet_name='Engine Timeline')
engine_df.columns = [c.lower().replace(' ','_') for c in engine_df.columns]

print(f"\nShape: {engine_df.shape}")
print(f"\nColumns: {list(engine_df.columns)}")
```

```python
conn = get_connection()
engine_df.to_sql('engines',conn,if_exists='replace',index=False)
conn.commit()
conn.close()
print("Engine Timeline imported Successfully")
```

```python
conn = get_connection()
print(pd.read_sql_query("SELECT COUNT(*) as total FROM engines", conn))
print(pd.read_sql_query("SELECT * FROM engines LIMIT 3", conn))
conn.close()
```
